```python
#Corrigido 
class Livro:
    def __init__(self, titulo, autor):
        self._titulo = titulo
        self._autor = autor
        self._esta_emprestado = False

    @property
    def titulo(self):
        return self._titulo

    @property
    def autor(self):
        return self._autor

    @property
    def esta_emprestado(self):
        return self._esta_emprestado

    def emprestar(self):
        if self._esta_emprestado:
            raise Exception(f'O livro "{self._titulo}" já está emprestado.')
        self._esta_emprestado = True

    def devolver(self):
        if not self._esta_emprestado:
            raise Exception(f'O livro "{self._titulo}" não está emprestado.')
        self._esta_emprestado = False


class Biblioteca:
    def __init__(self):
        self._acervo = []

    def adicionar_livro(self, titulo, autor):
        try:
            if self._livro_existe(titulo):
                raise Exception(f'O livro "{titulo}" já está cadastrado no acervo.')
            novo_livro = Livro(titulo, autor)
            self._acervo.append(novo_livro)
            print(f'Livro "{titulo}" adicionado ao acervo.')
        except Exception as e:
            print(f'Erro ao adicionar livro: {e}')

    def emprestar_livro(self, titulo):
        try:
            livro = self._buscar_livro(titulo)
            if livro.esta_emprestado:
                raise Exception(f'O livro "{titulo}" já está emprestado.')
            livro.emprestar()
            print(f'Livro "{titulo}" foi emprestado.')
        except Exception as e:
            print(f'Erro ao emprestar livro: {e}')

    def devolver_livro(self, titulo):
        try:
            livro = self._buscar_livro(titulo)
            if not livro.esta_emprestado:
                raise Exception(f'O livro "{titulo}" não está emprestado.')
            livro.devolver()
            print(f'Livro "{titulo}" foi devolvido.')
        except Exception as e:
            print(f'Erro ao devolver livro: {e}')

    def _buscar_livro(self, titulo):
        for livro in self._acervo:
            if livro.titulo == titulo:
                return livro
        raise Exception(f'Livro "{titulo}" não encontrado no acervo.')

    def _livro_existe(self, titulo):
        return any(livro.titulo == titulo for livro in self._acervo)

# Exemplo de uso:
if __name__ == "__main__":
    biblioteca = Biblioteca()
    biblioteca.adicionar_livro("Dom Casmurro", "Machado de Assis")
    biblioteca.adicionar_livro("Dom Casmurro", "Machado de Assis")  # Teste de duplicidade
    biblioteca.emprestar_livro("Dom Casmurro")
    biblioteca.emprestar_livro("Dom Casmurro")  # Teste de já emprestado
    biblioteca.devolver_livro("Dom Casmurro")
    biblioteca.devolver_livro("Dom Casmurro")  # Teste de devolução não
```
